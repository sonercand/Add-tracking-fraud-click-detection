```python
import pandas as pd
import numpy as np
import random
import os
```

#### merge pickle files

```python
train_files = os.listdir('.\processed_train_parts1')
```

```python
frames = []
data1 = pd.DataFrame()
for l in np.arange(len(train_files)):
    if l%5 ==0:
        print(l)
    data = pd.read_pickle('.\processed_train_parts1\part'+str(l+1)+'.pkl')
    frames =[data1,data]
    data1 = pd.concat(frames)
```

```python
del frames
del data
del train_files
del l
```
